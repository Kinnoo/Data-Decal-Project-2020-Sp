## Calculate the increase rate or decrease rate of stock price

```python
def change(column):
    #buy price
    buyPrice=column[0]
    #current price
    
    curPrice=column[column.size-1]
    #accumulated change
    priceChange=(curPrice-buyPrice)/buyPrice
    
    if(priceChange>0):
        print('accumulated increase rate =',priceChange*100,'%')
    elif(priceChange==0):
        print('no change=',priceChange*100,'%')
    else:
        print('accumulated decrease rate = ',priceChange*100,'%')

    return priceChange
```

```python
closeCol=babaDf['Close']
babaChange=change(closeCol)
```
